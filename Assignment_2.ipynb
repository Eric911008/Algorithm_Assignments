```python
import time
import matplotlib.pyplot as plt
import numpy as np

# Top-down approach
def fibonacci_recursive(n):
    if n <= 1:
        return n
    else:
        return fibonacci_recursive(n-1) + fibonacci_recursive(n-2)

# Bottom-up approach
def fibonacci_dynamic(n):
    fib = [0] * (n+1)
    fib[1] = 1
    for i in range(2, n+1):
        fib[i] = fib[i-1] + fib[i-2]
    return fib[n]

# Task 1: Measure execution time of F(1) to F(100) using both methods
n_values = np.arange(1, 101)
recursive_times = []
dynamic_times = []

for n in n_values:
    # Measure execution time for top down approach
    try:
        start_time = time.time()
        fibonacci_recursive(n)
        end_time = time.time()
        recursive_times.append(end_time - start_time)
    except KeyboardInterrupt:
        print(f"Stopped at n = {n}")
        break
    
    # Measure execution time for bottom up approach
    try:
        start_time = time.time()
        fibonacci_dynamic(n)
        end_time = time.time()
        dynamic_times.append(end_time - start_time)
    except KeyboardInterrupt:
        print(f"Stopped at n = {n}")
        break

# Plotting execution times
plt.figure(figsize=(10, 6))
plt.plot(n_values[:len(recursive_times)], recursive_times, label='Top-down')
plt.plot(n_values[:len(dynamic_times)], dynamic_times, label='Bottom-up')
plt.xlabel('n')
plt.ylabel('Execution Time (s)')
plt.title('Execution Time of Fibonacci Calculation')
plt.legend()
plt.grid(True)
plt.show()

# Task 2: Measure the degree of overlapping subproblems
fib_4_times = []

for i in range(5, 51):
    try:
        start_time = time.time()
        fibonacci_recursive(4)
        end_time = time.time()
        fib_4_times.append(end_time - start_time)
    except KeyboardInterrupt:
        print(f"Stopped at n = {i}")
        break

# Plotting the degree of overlapping subproblems
plt.figure(figsize=(10, 6))
plt.plot(range(5, 5+len(fib_4_times)), fib_4_times)
plt.xlabel('Value of n')
plt.ylabel('Execution Time of F(4) (s)')
plt.title('Degree of Overlapping Subproblems')
plt.grid(True)
plt.show()
```
